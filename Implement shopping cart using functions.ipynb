```python
Shoplist=[]
rem=[]
cost = 0
total = 0
import random

def add(): 
    global cost
    item=str(input("Add item :"))
    price=int(input("Enter  cost of item "))
    Shoplist.append(item)
    cost=(cost)+(price)

def remove():
    global cost
    item=str(input("Remove item ::"))
    price=int(input("Enter  cost of item "))
    rem.append(item)
    Shoplist.remove(item)
    cost=(cost)-(price)
    
def rebuy():
    global cost
    print(rem)
    item=str(input("rebuy item : "))
    price=int(input("Enter  cost of item "))
    Shoplist.append(item)
    cost=(cost)+(price)
    
def discount():
    global cost
    global total
    total=int(cost)
    discount=input('Do you have any coupon code? Y or N')
    while discount.lower() == "y":
        coupon=random.randint(1, 5)
        choice=int(input('Enter your Coupon Number btw 1 to 5'))
        if int(coupon) == int(choice):
            print("dicount\n")
            total=(cost) * 0.1
            cost=(cost)-(total)
        else:
            print("invalid coupone code\n")
            total=int(cost)
        break
        
def generate():
    discount()
    global cost 
    global total
    print("Here is your shopping list.")
    for listitem in Shoplist:
        print(listitem)
    
    print('Your total bill is'+ ' ' + str(total) +' ' + 'Rs.')
    
while(True):
    print("\n 1. add item\n",'2. remove item\n','3. rebuy\n','4. generate bill\n','5. Exit\n')
    ch=int(input('Enter choice: '))
    if ch==1:
        add()
    elif ch==2:
        remove()
    elif ch==3:
        rebuy()
    elif ch==4:
        generate()
    elif ch==5:
        break
    else:
        print("Invalid Choice")
    
```
